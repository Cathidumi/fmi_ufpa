# ALGORITMO PROPHET

## Como funciona o Prophet:

O Prophet é especialmente útil para conjuntos de dados que:

    Contêm um período longo (meses ou anos) de observações históricas detalhadas.
    Têm várias sazonalidades fortes;
    Incluem eventos importantes conhecidos anteriormente, mas irregulares;
    Têm pontos de dados ausentes ou grandes exceções;
    Mostram tendências de crescimento não lineares que estão se aproximando de um limite.

O Prophet é um modelo de regressão aditiva com uma parte linear ou curva de tendência de crescimento logístico. Ela inclui um componente sazonal anual modelado usando série Fourier e um componente sazonal semanal modelado usando variáveis fictícias. Tendências não lineares são ajustadas com sua sazonalidade.
Prophet é robusto para dados ausentes e mudanças na tendência.
.
Automatizado 
Previsões ajustáveis
Intitivo para não especialistas
.
Modelo de regrassão aditivo, com 3 principais componentes: Tendência, sazonalidade e feriados.
	

```
# y(t)=g(t)+s(t)+h(t)+t

```

	g(t) => Tendência == mudanças não periódicas
	s(t) => Sazonalidade == mudanças periodica 
	h(t) => Efeito de feriados == maiores mudanças
	t => Erro == ruido

O q o Prophet informa:

	tendencia
	Tendencia de baixa
	tendencia de alta
	'yhat'(Previsão em si)

````python
# Carregando arquivos
import pandas as pd
from fbprophet import Prophet

dados = pd.read_csv('dados_8_horas.csv', usecols=['INS', 'TEM_INS'], parse_dates=['INS'])
dados.rename(columns={'INS': 'ds', 'TEM_INS': 'y'}, inplace = True)
dados
````

````python
# Instânciando um novo objeto
m = Prophet()
m.fit(dados)

# Obter um dataframe que se estende até o futuro por um determinado número de dias. Por padrão, também incluirá as datas do histórico.
future = m.make_future_dataframe(periods=24)
future.tail()
````

````python
# O método predict atribuirá a cada linha em valor de previsão.
# O forecast é um novo dataframe que inclui uma coluna com a previsão, bem como colunas para componentes e intervalos de incerteza.
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
````

````python
# Exportar as últimas 24 linhas do novo .csv.
forecast.rename(columns={'yhat': 'TEM_INS', 'ds': 'INS'}, inplace = True)
forecast[-24:].to_csv('teste_saida.csv',  index=False, columns = ['INS', 'TEM_INS'])
````
