#Tratanto Index

```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('dados.csv', sep=';')
```

```python
dados['DAT_MED'] = dados['DT_MEDICAO']
dados['HR_MED'] = dados['HR_MEDICAO']
dados.pop('DT_MEDICAO')
dados.pop('HR_MEDICAO')
```

```python
dados.head()
```

```python
dados['HORA'] = dados['HR_MED'] / 100
dados['HORA'] = dados['HORA'].astype(np.int64)
dados.head()
```

```python
dados['INS'] = dados['DAT_MED'].astype(str) + ';' + dados['HORA'].astype(str)
dados
```

```python

dados['INS'] = pd.to_datetime(dados['INS'], format='%d/%m/%Y;%H')
dados.set_index('INS', inplace=True)
dados.head()
```

#Exportando os dados

```python
dados.to_csv('dados_new.csv', index = True)
```
