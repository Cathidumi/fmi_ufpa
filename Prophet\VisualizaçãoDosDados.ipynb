# Arquivo feito pra visualização dos dados dos últimos 5 dias até 07/01/2022



```python
import pandas as pd

anual = 'Dados um ano.csv'
mensal = 'Dados um mes.csv'

# Lê o .csv com os dados base
dados = pd.read_csv(mensal, usecols=['INS', 'HORA', 'TEM_INS', 'UMD_INS', 'VEN_VEL'], parse_dates=['INS'])
dados
```

```python
# Importar dados para um DataFrame
df = pd.read_csv(mensal, usecols=['INS', 'TEM_INS'], parse_dates=['INS'])
df.rename(columns={'INS': 'ds', 'TEM_INS': 'y'}, inplace = True)
```

```python
df.sort_values('ds', inplace=True)
df.set_index('ds', inplace=True)
df.plot()
```

```python
for c in range(24):
  teste = dados[dados['HORA'] == c]
  teste.to_csv('HORA/HORA'+str(c)+'.csv', index=False)
```

```python
for c in range(24):
  temp0 = pd.read_csv('HORA/HORA'+str(c)+'.csv', usecols=['INS', 'TEM_INS'], parse_dates=['INS'])
  temp0.rename(columns={'INS': 'ds', 'TEM_INS': 'y'}, inplace = True)

  temp0.sort_values('ds', inplace=True)
  temp0.set_index('ds', inplace=True)
  temp0.plot(title=c)
```
